```python
from sklearn.model_selection import train_test_split
# Splitting the dataset into training and testing sets 
```

```python
X = titanic.drop('Survived', axis = 1)
# Dropping the attribute which needs to predicted by the Logistic_Regression
```

```python
y = titanic['Survived']
# The label attribute i.e. the attribute which needs to be predicted whether the passenger had survived or not
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=42)
# 81% of the data for the training and building the model and 30% for testing and validation of the model
```

```python
from sklearn.linear_model import LogisticRegression
# Importing Logistic Regression from the Scikit learn package
```

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(X_train,y_train)
# Fitting the Logistic Regression with the training dataset
```

```python
predictions = logistic_model.predict(X_test)
# Predicting the label attribute i.e. survived from the titanic using Logistic Regression algorithm
```

```python
from sklearn.metrics import classification_report
# Importing the metrics for validating the model using the Scikit learn package and importing the confusion matrix and classification report
```

```python
print(classification_report(y_test,predictions))
# The classification report gives the comparsion between prediction result and original result
```
